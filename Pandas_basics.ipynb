### Pandas

1) Podstawy

```python
import pandas as pd
import numpy as np
```

```python
myData=np.arange(0,100)
myData
```

```python
myData=myData.reshape(10,10)
```

```python
myData
```

```python
# przypisanie tablicy do DataFrame
pd.DataFrame(myData)
```

```python
# Definiowanie nazw kolumn / wierszy
my_index = ['Row1','Row2','Row3','Row4','Row5','Row6','Row7','Row8','Row9','Row10']
my_col = ['Col1','Col2','Col3','Col4','Col5','Col6','Col7','Col8','Col8','Col10']
```

```python
df = pd.DataFrame(data=myData,index=my_index,columns=my_col)
```

```python
df
```
