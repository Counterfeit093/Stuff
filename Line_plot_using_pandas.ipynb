### Plot using pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
housing = data('Housing')
```

```python
#wykres pokazuje wszystkie kolumny typu liczbowego
housing.plot()
```

```python
#przypisanie wybranych kolumn do y pozwoli zwiekszyc widocznosc.
#" ; " na koncu polecenia ukryje dodatkowy komentarz
housing.plot(y='price');
```

```python
#teraz wykres wyglada znacznie lepiej
housing.plot(
    y='price',
    title='House Prices',
    xlabel='House Index', #opis osi
    ylabel='Price',
    legend=False, #Pokazuje legende
    figsize=(20,10) # rozmiar, 20 = x, 10 =y
);
```

```python
#zestawienie danych, dodanie kolumny lotsize
housing.plot(
    y=['price','lotsize'],
    title='House Prices and Lot Size',
    xlabel='House Index',
    legend=True,
    subplots = True # Dodaje kolejny wykres
);
```

```python
#zmiana pozycji wykresow wzgledem siebie
housing.plot(
    y=['price','lotsize'],
    title='House Prices and Lot Size',
    xlabel='House Index',
    legend=True,
    subplots = True,
    layout=(1,2)
);
plt.tight_layout() #wymaga matplotlib.pyplot, wyrownuje oba wykresy wzgledem siebie
```
